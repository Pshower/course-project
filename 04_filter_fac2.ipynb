```python
import pandas as pd
import numpy as np
from multiprocessing.dummy import Pool as ThreadPool
import time
```

```python
from q_tools.data import get_index_codes
```

```python
# 获取因子文件
def get_fac2_new2():
    data=pd.read_csv('01_data/fac2/fac_new2.csv',index_col=[0,1],parse_dates=['date'],dtype={'code': 'str'})
    return data
```

```python
# 获取对应股票的文件
def get_fac2_share(data,code):
    return data.loc[(slice(None),code),:]
```

```python
share_list=get_index_codes()
```

```python
fac2=get_fac2_new2()
fac2
```

```python
for i in range(len(share_list)):
    tmp=get_fac2_share(fac2,share_list[i])
    tmp.to_csv('01_data/trade_list_all/%s.csv'% (share_list[i]))
```

```python
code=share_list[0]
df=get_fac2_share(fac2,code)
for code in share_list[1:]:
    df1=get_fac2_share(fac2,code)
    df=pd.concat([df,df1])
```

```python
df.to_csv('01_data/trade_list_all/all289.csv')
```
